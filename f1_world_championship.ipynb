```python
import pandas as pd
import numpy as np
import requests
from tqdm import tqdm
```

```python
years=range(1950,2021)
years=list(years)
```

```python
url='https://en.wikipedia.org/wiki/{}_Formula_One_World_Championship'
url_fone='https://en.wikipedia.org/wiki/{}_Formula_One_season'
D={}
for y in tqdm(years):
    if y<1981:
        u=url_fone.format(y)
    else:
        u=url.format(y)
    data=pd.read_html(u)
    for d in data:
        b=(d=='Pos').sum().sum() and (d=='Driver').sum().sum()
        a=(d=='Pos.').sum().sum() and (d=='Driver').sum().sum()
        if a or b:
            D[y]=d
```

```python
for _,d in D.items():
    d.columns=d.iloc[0]
    d.drop(0,axis=0,inplace=True)
```

```python
for _,d in D.items():
    cols={}
    for col in d.columns:
        if col.find('Pos')!=-1:
            cols[col]='Pos'
        elif col.find('Driver')!=-1:
            cols[col]='Driver'
    d.rename(columns=cols,inplace=True)
```

```python
points=pd.DataFrame()
for y,d in D.items():
    f=d.iloc[:,:-1].copy()
    f['year']=y
    points=pd.concat([points,f],sort=True)
```

```python
cols=points.columns.tolist()
cols2=['year','Driver','Pos']
for c in cols:
    if c not in ['Pos','Driver','year']:
        cols2.append(c)
```

```python
points=points[cols2]
```

```python
points.sort_values(['year','Driver','Pos'],inplace=True)
```
